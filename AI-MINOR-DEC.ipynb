# Division of One Sex Column into 2 columns which are sex_0 & sex_1, which represents Female & Male respectively

# (1 = male; 0 = female)

```python
features=pd.get_dummies(features,columns=["sex"])
```

```python
features
```

# Specify the Architecture

```python
import keras
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical

```

```python
target=to_categorical(Y)
```

```python
target.shape
```

```python
features.shape
```

```python
model=Sequential()
```

```python
model.add(Dense(50,activation="relu",input_shape=(14,)))
model.add(Dense(100,activation="relu"))
model.add(Dense(2,activation="softmax"))
```

# Compile the Model

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

# Fit the Model

```python
model.fit(features,target,epochs=20)
```
